```python
# calculate mutiple statistics after grouping
# Example setup
df = pd.DataFrame({
    'Department': ['Sales', 'Sales', 'IT', 'IT', 'HR', 'HR'],
    'Salary': [50000, 60000, 70000, 80000, 45000, 50000],
    'Bonus': [2000, 3000, 5000, 4000, 1000, 1500]
})

# Grouping by 'Department' and calculating mean and max for 'Salary'
summary = df.groupby('Department')['Salary'].agg(['mean', 'max', 'count'])
print(summary)
```

```python
summary = df.groupby('Department').agg(
    total_staff=('Department', 'count'),
    average_salary=('Salary', 'mean'),
    max_bonus=('Bonus', 'max'),
    salary_range=('Salary', lambda x: x.max() - x.min())
)
print(summary)
```

## Confidence intervals and hypothesis testing

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats as stats
import statsmodels.api as sm
```

```python
employees = pd.DataFrame({
    "name": ["Alice Smith", "Bob Johnson", "Charlie Williams", 
             "Diana Brown", "Ethan Jones"],
    "age": [28, 45, 34, 29, 50],
    "job_title": ["Data Analyst", "Senior Developer", "Project Manager", "HR Specialist", "Director"],
    "tenure": [3, 10, 1, 5, 0]
})
employees.head()
```

```python
### START CODE HERE ###

# Calcuate the sample size 
n = employees.shape[0]

# Calculate the mean of age
mean_age = employees["age"].mean()

# Calculate the standard deviation of age
std_age = employees["age"].std()

# Calculate the standard error of the mean (SEM)
SEM = std_age / np.sqrt(n)

# Set the confidence level to 0.95
conf = 0.95

# Calculate the 95% confidence interval for the mean age
confidence_interval = stats.norm.interval(conf, loc=mean_age, scale=SEM)

### END CODE HERE ###

print(mean_age)
print(confidence_interval)
```
